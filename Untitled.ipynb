```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python

def fibonacci(n):
    li=[0,1]
    for i in range(2,n):
        a=li[-1]
        b=li[-2]
        li.append(a+b)
    return li
```

```python
Y=np.arange(-100,100,1)
Y.shape
```

```python
fig,axs=plt.subplots(figsize=(8,5))
for i in range(0,101,1):
    I=np.array([i])
    I=np.repeat(I,Y.shape)
    axs.fill(Y,Y**2+i**2)
    plt.fill(Y,I)
    axs.plot()
plt.show()
```
